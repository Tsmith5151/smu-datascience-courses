**List:**

Create a list named my_list in python with the following data points: [45.4, 44.2, 36.8,  35.1, 39.0, 60.0, 47.4 ,41.1, 45.8, 35.6]

   - a.) Print the 5th element in the list
   - b.) Append 55.2 to my_list
   - c.) Remove the 6th element in the list
   - d.) Iterate over the list to print data points greater than 45


```python
#Create list:
my_list = [45.4, 44.2, 36.8,  35.1, 39.0, 60.0, 47.4 ,41.1, 45.8, 35.6]

#print 5th element
print "The fifth element in the list is: %s \n" % (my_list[4])

#Append list
my_list.append(55.2)
print "Appended List: \n %s" % (my_list)

#Remove the 6th element
print "\nRemove the 6th element from list: %s\n" %(my_list.pop(5))

#Loop through list and print the data points greater than 45
print "New List with data points in my_list greater than 45:"
new_list = []
for i in my_list:
    if i >= 45:
        new_list.append(i)
print new_list
```

**Introduction to numpy:**
   - a.) Import the numpy library
   - b.) Declare numpy array with the same data points as in my_list
   - c.) Compute the mean and standard deviation of my_list
   - d.) Use logical referencing to get only those values that are less than 45
   - e.) Compute the max and min of the array using numpy.max() and numpy.min()


```python
#import numpy
import numpy as np

#Array
my_array = np.array(my_list)
print "my_arrary = %s \n" % (my_array)

#my_list' Statistics
print "Statistics:"
print "Mean: %s" %(np.mean(my_array))
print "Standard Deviation: %s" %(round(np.std(my_array),3))
print "Maximum: %s" % np.max(my_array)
print "Minimum: %s" % np.min(my_array)

#Get values that are less than 45 using comprehension:
my_list_comp = [round(x,3) for x in my_array if x<=45]
print "\nData points less than 45: \n %s" % (my_list_comp)
```
